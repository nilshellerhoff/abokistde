```python
from collections import deque
points1 = [
    '8,8', '48,8', '88,8',
    '88,48', '48,48', '8,48',
    '8,88', '48,88', '88,88'
]
points2 = []
for point in points1:
    points2.append(point)
    points2.append(point)

times = []
for i,p in enumerate(points2):
    times.append(i/len(points2))
times.append(1)
timeStr = "; ".join([str(t) for t in times])

bezier = '1 0 0 1'
bezierStr = "; ".join([bezier for i in range(len(points2))])

def rect(color='#e2e5ea', startingPoint='88,88', duration=9):
    startindex = points2.index(startingPoint)
    points = deque(points2)
    points.rotate(-1*startindex-1)
    # append final point
    points.append(startingPoint)

    values = "; ".join(points)
    rect = """
<rect
    style="fill:{color};fill-opacity:1;stroke:#666666;stroke-width:0"
    id="rect833"
    width="32"
    height="32"
    x="0"
    y="0">
    <animateMotion
        dur="{duration}s" 
        values="{values}" 
        keyTimes="{times}"
        repeatCount="indefinite"
        calcMode="spline"
        keySplines="{bezier}"/>
</rect>
    """.format(color=color, values=values, times=timeStr, bezier=bezierStr, duration=duration)

    return rect
```

```python
rects = [
    ('#e2e5ea', '8,8'),
    ('#777b84', '48,8'),
    ('#635da7', '88,8'),
    ('#00a5dd', '8,48'),
    ('#93c937', '48,48'),
    ('#ffe001', '88,48'),
    ('#f68f18', '8,88'),
    ('#dc2031', '48,88'),
    ('#dc3e93', '88,88')
]

rstr = ""
import clipboard

for r in rects:
    rstr += rect(color=r[0], startingPoint=r[1])

clipboard.copy(
    svg.format(rectangles=rstr)
)
```
